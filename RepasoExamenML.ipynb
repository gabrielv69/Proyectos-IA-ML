```python
df['TernureAge']=df.Tenure/df.Age
df['BalanceSalary']=df.Balance/df.EstimatedSalary
df['CreditAge']=df.CreditScore/df.Age
```

```python
df_var_cont=['CreditScore','Age', 'Tenure' , 'Balance','NumOfProducts', 'EstimatedSalary','TernureAge', 'BalanceSalary','CreditAge' ] 
df_var_cat=['Geography','Gender', 'HasCrCard','IsActiveMember']
```

```python
df_aux= df[['Exited']+df_var_cont+df_var_cat]
```

```python
df_aux
```

```python
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
```

```python
le = LabelEncoder()
df_aux['Gender']=le.fit_transform(df_aux['Gender'])
df_aux['Geography']=le.fit_transform(df_aux['Geography'])
```

```python
df_aux
```

```python
scaler = MinMaxScaler()
df_aux[df_var_cont]=scaler.fit_transform(df_aux[df_var_cont])
```

```python
df_aux
```

```python
from sklearn.model_selection import train_test_split
```

```python
X =df_aux.drop('Exited',axis=1)
y= df_aux[ 'Exited']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(12,activation='relu'))
model.add(Dense(6,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer = 'adam', loss= 'binary_crossentropy',metrics= [ 'accuracy'])
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test.values),batch_size=40,epochs=80,verbose=1)
```

```python
loss= pd.DataFrame( model.history.history)
plt.figure(figsize=(15,10))
sns.lineplot(data=loss,lw=3)
sns.despine()
```

```python
svm.score(X_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier # Sueldo Regression
from sklearn.metrics import accuracy_score
```

```python
error=[]
for n in range(1,50):
  knn = KNeighborsClassifier(n_neighbors=n)
  knn.fit(X_train,y_train)
  predic_n = knn.predict(X_test)
  error.append(np.mean(predic_n != y_test))
```

```python
plt.figure(figsize=(15,15))
plt.plot(range(1,50),error,color='g',linestyle='dashed',marker='x',markerfacecolor ='r')
```

```python
knn_c = KNeighborsClassifier(n_neighbors=8)
knn_c.fit(X_train,y_train)
```

```python
y_pred_Train = knn_c.predict(X_train)
y_pred_Test = knn_c.predict(X_test)
```

```python
print(accuracy_score(y_train,y_pred_Train))
print(accuracy_score(y_test,y_pred_Test))
```
